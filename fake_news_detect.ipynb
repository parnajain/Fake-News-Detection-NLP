```python
import string
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

Feature engineering-> combining the 'title' and 'text' column for simplified vectorisation

```python
df['text']= df['text'].str.replace(r'^.*?\(Reuters\)\s*-\s*','',regex=True)
```

```python
df['data'] = df['title'] + " " + df['text']
```

```python
df.drop(['title','text'], axis=1, inplace=True)
```

```python
df.head()
```

adding a new feature, that is the length of news article

```python
df['length'] = df['data'].apply(len)
df.head()
```

```python
sns.displot(x= 'length', col='Id',data= df, kde=True)
```

the feature 'length' is not really useful, hence we drop it

```python
df.drop('length', inplace=True, axis=1)
```

Text cleaning and preprocessing

```python
stop_words = set(stopwords.words('english'))
lemmatizer= WordNetLemmatizer()
```

```python
def text_preprocess(message):
    message = message.lower()
    message = re.sub(r"http\S+|www\S+|https\S+", "", message)

    no_punc = [char for char in message if char not in string.punctuation]
    no_punc = ''.join(no_punc)

    text=[]
    for word in no_punc.split():
        if word not in stop_words:

            lemma= lemmatizer.lemmatize(word)
            text.append(lemma)

    return ' '.join(text)
```

```python
df['data'] = df['data'].apply(text_preprocess)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

Using logistic regression model

```python
pipeline= Pipeline([
    ('bow', CountVectorizer(ngram_range=(1,2))),
    ('tfidf', TfidfTransformer()),
    ('lr', LogisticRegression())
])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
msg_train, msg_test, label_train, label_test= train_test_split(df['data'], df['Id'], test_size=0.3)
```

```python
pipeline.fit(msg_train, label_train)
```

```python
prediction= pipeline.predict(msg_test)
```

```python
print(classification_report(label_test, prediction))
print(confusion_matrix(label_test, prediction))
```

using naive bayes 

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
pipelineNB= Pipeline([
    ('bow', CountVectorizer(ngram_range=(1,2))),
    ('tfidf', TfidfTransformer()),
    ('nb', MultinomialNB())
])
```

```python
pipelineNB.fit(msg_train, label_train)
```

```python
predictionNB= pipeline.predict(msg_test)
```

```python
print(classification_report(label_test, predictionNB))
print(confusion_matrix(label_test, prediction))
```

Naive Bayes and Logistic Regression are giving the same results as they are both linear classifiers.

Using 5-fold cross validation with F1 score for unbiased results

```python
from sklearn.model_selection import cross_val_score
```

Using a DummyClassifier with the most frequent strategy as a baseline to ensure that the NLP 
models were genuinely learning meaningful patterns and not benefiting from dataset bias or leakage.

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy='most_frequent')
dummy_scores = cross_val_score(dummy, df['data'], df['Id'], cv=5, scoring='f1')
print(dummy_scores.mean())
```

Since the dataset is imbalanced, using F1 score is better than accuracy

```python
scores = cross_val_score(pipeline, df['data'], df['Id'], cv=5, scoring='f1')
print(scores)
```

```python
scores2 = cross_val_score(pipelineNB, df['data'], df['Id'], cv=5, scoring='f1')
print(scores2)
```

```python
print("Mean F1 score for logistic regression: ", scores.mean())
print("Mean F1 score for naive-bayes: ", scores2.mean())
```

using cross-validation, we see that logistic regression outperforms the naive-bayes model

Naive Bayes assumes every word is independent.Logistic Regression can handle the relationship between words better.

Interpreted model predictions using Logistic Regression feature coefficients to identify linguistically meaningful indicators of fake and real news

```python
lr_coeff = pipeline.named_steps['lr'].coef_[0]
feature_names= pipeline.named_steps['bow'].get_feature_names_out()
coeff_df= pd.DataFrame({'Feature': feature_names,
                       'Coefficient': lr_coeff})
```

```python
print(coeff_df)
```

```python
top_false = coeff_df.sort_values(by='Coefficient', ascending=False).head(10)
top_true= coeff_df.sort_values(by='Coefficient').head(10)
```

```python
print(top_true)
```
